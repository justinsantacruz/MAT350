```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.where(abs(x) < np.pi/2, 1 + 1j, np.where(x > np.pi/2, -1j, 0))

def fourier_series(x, n=5):
    a0 = np.pi/4 + 1j*np.pi/4
    an = lambda n: 1j*np.sin(n*np.pi/2) + np.cos(n*np.pi/2)
    bn = lambda n: (1 - (-1)**n)/(n*np.pi) + 1j*(-1)**(n+1)/(n*np.pi)
    sum = a0
    for i in range(1, n+1):
        sum += (an(i) + bn(i)*1j) * np.exp(1j*i*np.pi*x/2)
    return sum.real, sum.imag

x = np.linspace(-2, 2, 500)
y_real, y_imag = fourier_series(x, n=5)

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
axs[0, 0].plot(x, y_real, label='Series')
axs[0, 0].plot(x, f(x).real, label='Function')
axs[0, 0].legend()
axs[0, 0].set_title('Real part of function and series')

axs[0, 1].plot(x, y_imag, label='Series')
axs[0, 1].plot(x, f(x).imag, label='Function')
axs[0, 1].legend()
axs[0, 1].set_title('Imaginary part of function and series')

y_real, y_imag = fourier_series(x, n=1)

axs[1, 0].plot(x, y_real, label='Series')
axs[1, 0].plot(x, f(x).real, label='Function')
axs[1, 0].legend()
axs[1, 0].set_title('Real part of function and series (n=1)')

axs[1, 1].plot(x, y_imag, label='Series')
axs[1, 1].plot(x, f(x).imag, label='Function')
axs[1, 1].legend()
axs[1, 1].set_title('Imaginary part of function and series (n=1)')

plt.show()
```
